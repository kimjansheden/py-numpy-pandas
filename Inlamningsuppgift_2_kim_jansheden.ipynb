## Part 1 - Numpy

```python
import numpy as np
```

```python
# 1
# Check if array contains duplicates.
def contains_duplicates(array):
    seen = []

    def contains_duplicates(array):
        if array.ndim == 1:                     # Unpack the array until there is only 1 dimension left.
            for element in array:               # Loop through all dimensions.
                if seen.count(element) == 1:    # If the element already exists:
                    return True                 # then the array contains duplicates.
                seen.append(element)
        else:                                   # If there are more than 1 dimensions
            for dimension in array:                     # unpack it
                if contains_duplicates(dimension):      # and call the function again recursively until a 1 dimensional array is found.
                    return True                         # If any duplicates are found in this dimension.
        return False
    
    return contains_duplicates(array)           # Unroll the recursion again and return the boolean result.

arr = np.random.uniform(0.0,10.0,(5,10))
print(arr)
print(arr.ndim)
print(arr.shape)
print(arr.size)
print(arr.dtype)

dup = contains_duplicates(arr)
print(f"Does the array contain duplicates? {dup}")
if not dup:
    print("All values are unique.")
```

Concatenate the ages to be as the Persons suspected of offences data

Let's see what we've got.

```python
pop_ages.head(20)
```

```python
# Rename columns, group and sum by age.
# Concatenate the ages and delete the old columns.

pop_ages.rename({"kön": "Kön", "2021": "Antal"}, inplace=True, axis=1)
pop_ages = pop_ages.groupby(pop_ages["Ålder"]).sum().reset_index().sort_index(axis=1, ascending=False)

# Join together the ages like the Persons suspected of offences dataframe.

# Create new rows with the sums of the ages to be concatenated.
new_row_21 = pd.DataFrame({'Ålder': ["21-24"], 'Antal': [pop_ages[pop_ages["Ålder"].between(21, 24)]["Antal"].sum()]})
new_row_25 = pd.DataFrame({'Ålder': ["25-29"], 'Antal': [pop_ages[pop_ages["Ålder"].between(25, 29)]["Antal"].sum()]})
new_row_30 = pd.DataFrame({'Ålder': ["30-39"], 'Antal': [pop_ages[pop_ages["Ålder"].between(30, 39)]["Antal"].sum()]})
new_row_40 = pd.DataFrame({'Ålder': ["40-49"], 'Antal': [pop_ages[pop_ages["Ålder"].between(40, 49)]["Antal"].sum()]})
new_row_50 = pd.DataFrame({'Ålder': ["50-59"], 'Antal': [pop_ages[pop_ages["Ålder"].between(50, 59)]["Antal"].sum()]})
new_row_60 = pd.DataFrame({'Ålder': ["60-"], 'Antal': [pop_ages[pop_ages["Ålder"].between(60, 110)]["Antal"].sum()]})

# Find the index where to put the new rows.
idx_20 = pop_ages[pop_ages["Ålder"] == 20].index[0] +1
idx_25 = pop_ages[pop_ages["Ålder"] == 25].index[0] +1
idx_30 = pop_ages[pop_ages["Ålder"] == 30].index[0] +1
idx_40 = pop_ages[pop_ages["Ålder"] == 40].index[0] +1
idx_50 = pop_ages[pop_ages["Ålder"] == 50].index[0] +1
idx_60 = pop_ages[pop_ages["Ålder"] == 60].index[0] +1

# Drop the old rows of Ålder.
pop_ages = pop_ages[~pop_ages["Ålder"].between(21, 24)].reset_index(drop=True)
pop_ages = pop_ages[~pop_ages["Ålder"].between(25, 29)].reset_index(drop=True)
pop_ages = pop_ages[~pop_ages["Ålder"].between(30, 39)].reset_index(drop=True)
pop_ages = pop_ages[~pop_ages["Ålder"].between(40, 49)].reset_index(drop=True)
pop_ages = pop_ages[~pop_ages["Ålder"].between(50, 59)].reset_index(drop=True)
pop_ages = pop_ages[~pop_ages["Ålder"].between(60, 110)].reset_index(drop=True)

# Insert the new rows in Ålder at the right places.
pop_ages = pd.concat([pop_ages.iloc[:idx_20], new_row_21, pop_ages.iloc[idx_20:]]).reset_index(drop=True)
pop_ages = pd.concat([pop_ages.iloc[:idx_25], new_row_25, pop_ages.iloc[idx_25:]]).reset_index(drop=True)
pop_ages = pd.concat([pop_ages.iloc[:idx_30], new_row_30, pop_ages.iloc[idx_30:]]).reset_index(drop=True)
pop_ages = pd.concat([pop_ages.iloc[:idx_40], new_row_40, pop_ages.iloc[idx_40:]]).reset_index(drop=True)
pop_ages = pd.concat([pop_ages.iloc[:idx_50], new_row_50, pop_ages.iloc[idx_50:]]).reset_index(drop=True)
pop_ages = pd.concat([pop_ages.iloc[:idx_60], new_row_60, pop_ages.iloc[idx_60:]]).reset_index(drop=True)

print("The ages joined together to be consistent with the rest of the data".title())
pop_ages.head(20)
```

```python
# Select rows from the "brott" dataframe where the value of the "Brottstyp" column is not equal to "SAMTLIGA BROTT".
# And select the columns "Brottstyp" and ages.
top_5_crimes_per_age = brott.loc[brott["Brottstyp"] != "SAMTLIGA BROTT"]
top_5_crimes_per_age = top_5_crimes_per_age.iloc[:, [0] + list(range(2, 14))]

# Select only the rows where the "Brottstyp" column contains the string "§"
top_5_crimes_per_age = top_5_crimes_per_age[top_5_crimes_per_age.Brottstyp.str.contains("§").values]

# Display the filtered dataframe
print("Filtered dataframe with all crimes by age".title())
top_5_crimes_per_age
```

##### Top 5 Crimes By Age

Analyze the dataset that contains information on crime by age group. The objective is to extract the top 5 crimes for each age group, and present the results in a single dataframe. To achieve this, the code loops through each age group and processes the data for each group. The processed data for each group is then concatenated into a single dataframe.

```python
# Extract the columns for the different age groups.
age_columns = top_5_crimes_per_age.columns[1:]

# Creates an empty list that will contain the dataframes with the top 5 crimes for each age.
result = []

# Iterate through each age group.
for age in age_columns:
    # Create a new column name by replacing the "-" in the original column name with "_". This is because names with
    # a dash proved to cause trouble with the following operations.
    temp_age_name = age.replace("-", "_")
    
    # Select the columns "Brottstyp" and the current age group and add them to a new dataframe.
    # Rename the column for the current age group to the temp age name without dashes in order for nlargest to work.
    # Get the top 5 crimes for the current age group by number of occurrences.
    crimes = top_5_crimes_per_age[["Brottstyp", age]].rename(columns={age: temp_age_name}).set_index("Brottstyp").nlargest(5, temp_age_name)
    
    # Reset the index and add a column for the current age group.
    crimes = crimes.reset_index()
    crimes["Ålder"] = temp_age_name

    # Rename the column for the number of occurrences to "Antal"
    crimes = crimes.rename(columns={temp_age_name: "Antal"})

    # Append the processed data for the current age group to the `result` list.
    result.append(crimes)

# Concatenate all the processed data into a single dataframe.
top_5_crimes_per_age_all = pd.concat(result)

# Replace the "_" in the "Ålder" column back to "-".
top_5_crimes_per_age_all["Ålder"] = top_5_crimes_per_age_all["Ålder"].replace("_", "-", regex=True)

# Add 1 to the index of the dataframe for clarity in the output.
top_5_crimes_per_age_all.index = top_5_crimes_per_age_all.index + 1

# Display the final result.
top_5_crimes_per_age_all
```

##### Single Top Crime By Age

```python
# Replace the "-" in the original column name with "_". This is because names with a dash proved to cause trouble.
top_5_crimes_per_age_all["Ålder"] = top_5_crimes_per_age_all["Ålder"].replace("-", "_", regex=True)

# Group the data by the "Ålder" column. The lambda function within the groupby selects the row with the maximum value of "Antal" in each group.
top_5_crimes_per_age_all = top_5_crimes_per_age_all.groupby("Ålder").apply(lambda x: x.loc[x["Antal"].idxmax()])

# Reset the index of the dataframe, dropping the old index.
top_5_crimes_per_age_all = top_5_crimes_per_age_all.reset_index(drop=True)

# Create a mapping dataframe for the ages, used for sorting the data later.
df_mapping = pd.DataFrame({
    'Ålder': ['15', '16', '17', '18', '19', '20', '21_24', '25_29', '30_39', '40_49', '50_59', '60_'],
})

# Reset the index of the mapping dataframe and set it to be indexed by "Ålder".
sort_mapping = df_mapping.reset_index().set_index('Ålder')

# Create a new column in the main dataframe, "Ålder_num", that maps the values in the "Ålder" column to their index in the mapping dataframe.
top_5_crimes_per_age_all['Ålder_num'] = top_5_crimes_per_age_all['Ålder'].map(sort_mapping['index'])

# Sort the dataframe based on the "Ålder_num" column
top_5_crimes_per_age_all.sort_values("Ålder_num", inplace=True)

# Drop the "Ålder_num" column from the dataframe.
top_5_crimes_per_age_all.drop("Ålder_num", axis=1, inplace=True)

# Replace the "_" in the "Ålder" column back to "-".
top_5_crimes_per_age_all["Ålder"] = top_5_crimes_per_age_all["Ålder"].replace("_", "-", regex=True)

# Set the "Ålder" column as the index of the dataframe.
top_5_crimes_per_age_all.set_index("Ålder", inplace=True)

# Remove the name of the columns.
top_5_crimes_per_age_all.columns.name = ""

# Display the final result.
top_5_crimes_per_age_all
```
